```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
```

```python
data = load_digits()
dir(data)
```

```python
data.data[0]
```

```python
plt.gray()
plt.matshow(data.images[0])
```

```python
plt.gray()
for i in range(5):
    plt.matshow(data.images[i])
    plt.title(data.target_names[i])
    
```

### Splitting to Training and Testing datasets

```python
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2)
```

### Model training and tuning

```python
model = SVC(C=10)
```

```python
model.fit(X_train, y_train)
```

### Measuring accuracy score

```python
model.score(X_test, y_test)
```

### Highest accuracy score
    0.9861111111111112

```python
model.predict([data.data[1]])
```

```python
plt.gray()
for i in range(5):
    plt.matshow(data.images[i])
    plt.title(f"Actual {data.target[i]}")
    plt.xlabel(f"Predicted {int(model.predict([data.data[i]]))}")
```

```python
predict = model.predict(X_test)
correct_count = 0
wrong_count = 0
for i in range(len(predict)):
    print(f"Actual {y_test[i]} Predicted {predict[i]}")
    if y_test[i] == predict[i]:
        correct_count += 1
    else:
        wrong_count += 1
print(f"Correct prediction {correct_count}")
print(f"Wrong prediction {wrong_count}")        
```
