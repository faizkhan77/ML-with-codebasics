# Ensemble Learning

Lets say u want to buy an iphone but u wasnt sure if u should be buying it or not, 

So u then called 4 of your friends who already have an iphone and then ask for their opinion, 3 of them told u that u should buy it and 1 guy said u shouldn't buy it, so here u can take a Majority vote and end up buying the iphone since 3 people said to buy it vs 1 who said dont.

As u can see, We use ensemble learning in real life where to make a decision we take opinions from different people and do a majority vote, similarly in ML, sometimes what happen is if u have just 1 model and if u train that model using all ur dataset, that model might get overfit or it might suffer from a **High Variance**, so to take this **High Variance** problem, we can use **Ensemble Learning**.

In **Ensemble Learning** we train multiple models on the same dataset and when we do predictions, we do prediction on multiple models and whatever output we get, we combine that result somehow to get a final result

<b>Bagging</b> and <b>Boosting</b> are the two main techniques used in <b>Ensemble Learning</b> and today we are going to talk about <b>Bagging</b>!

# Bagging

Lets say we have this dataset of 100 samples

and when we train a ML model, one of the problem that we might encounter is overfitting and overfitting happens due to the nature of the datsaset, your ML methodology, etc

Usually overfitting model has a problem of **High Variance**.

To tackle this problem, one of the things we can do is, out of the 100 samples, create a small dataset of lets say 70 samples and the way you create this subset is by using **Resampling with Replacement**

now what is that exactly? so lets first understand that...

# Resampling with Replacement

Lets say we have this 10 samples out of which we want to create smaller dataset with 4 samples

In **Resampling with Replacemnet** we randmly pick any data point, lets say 4

And then, when we go and pick the second data point, we again randomly pick any datapoint from the main 1 to 10 dataset with equal probability and we dont look at what we already have in our subset.

So lets say second it picked 8

Same goes for the third, we randomly pick any datapoint from the main sample, and this time lets say it picked 4 again which is the same sample we already have in our subset and then we pick 9

Basically this is what we call **Resampling with Replacemnet** where in our subset we can get same data sample multiple times

So getting back to what we were doing, from our main dataset we created subset of 70 samples 

and we might create n number of such smaller subsets from our original dataset using  **Resampling with Replacemnet**

And then on individual dataset we can Train our ML model

Lets say we are trying to classify if a Person should buy insurance or not and lets say we are using **Logistic Regression**, so we will use **Logistic Regression** model, here M1, M2 and M3, they all are **Logistic Regression** but they are Trained on a different dataset and when they are trained, we have to perform the Prediction, We will Perform that Prediction on all 3 models in Parrallel individually and whatever result we get, we just take a Majority Vote

So above, M1 and M3 models are saying the Person should buy insurance and M2 is saying they shuldn't, hence our majority vote is 1 and that is our final outcome.

The benefit u get here is, these individual models (M1,M2,M3) are weak learners, Weak Learners means they are Trained on a subset of dataset and hence they will not overfit, U know its likely they will generalize better, they will not Overfit and these Individual subsets weak learners, when u Combine the result, u get an Overall Good Quality result.

This was the Case of **Classification**, same thing applies for **Regression**, Lets say you are doing House price prediction

You just take Average of whatever is the prediction by Individual model.

Now, this Technique is also called **Bootstrap Aggregation**

# Bootstrap Aggregation

It is called **Bootstrap Aggregation** cuz when u are creating these small subsets of datasets using **Resampling with Replacement**, that procedure is called **Bootstrap**

And when u combine their results using either an Average or Majority vote, that is called **Aggregation**



So hence, **Bagging** is also called **Bootstrap Aggregation**

# Random Forest Algorithm



Now **Random Forest Algorithm** is one of the **Bagging** Technique with only one Distinction, which is, Here we not only sample our Data rows but also samples our Features,

So basically, you sample your Rows as well as your Columns.

### House price dataset

Lets look at our Classic House price dataset where Town, Area, Bedrooms, etc are the Features and Price column is the Target Variable

Here, we will sample Rows and Columns both, so below u can see in the first Subset, we dont have a Bedrooms Column or a Plot Column cuz we Randomly Resample out of these 7 Columns (i.e. Town, Area, Bedrooms, Bathroom, Plot, School Rating and Basement)

So we pick only 4 Random columns and rows

Now, in the Second subset, again it Randomly sample it, so in this Subset we didnt get Area and Bedrooms column

And similarly, we do the same for the Third subset by Randomly picking the rows and columns both


So we are Resampling, we are Randomly picking rows as well as columns then on Individual subsets, we Train a **Decision Tree** model and then we just **Aggregate** the Results


Here we have **Decision Tree Regression** since we are doing **Regression** on House price dataset, 

But u can use **Decision Tree Classification** for **Classification** problem as well.

But the point is Very simple, **Random Tree** is basically a **Bagging** Technique but here we just do one Addiditonal thing which is we also Randomly pick the Features/Columns as well instead of just the Data rows, 

In other **Bagging** Technique maybe you only Randomly pick Data rows but in **Random Tree** we pick both Data rows and Features as well

# Bagging VS Bagged Trees

The difference between these terms **Bagging** and **Bagged Trees** is that,

1. In **Bagging**, Individual model can be any other models such as **SVM, KNN, Logistic Regression**, etc pretty much any models
2. Whereas in **Bagged Trees**, So **Random Forest** is a **Bagged Tree** example, Here, every model that u are Training is a Tree
