# Bayes Theorem

Thomas Bayes gave this famous equation of finding a conditional probability where u can find probability of A given the event B has occured by knowing some parameters which is the individual probabilities of A and B and knowing the probaility of B given that A has occured

So lets look at it in the context of our queen and diamond problem 

So here, this is the same equation we have represented for our specific use case. If u put them into our equation, u can easily find this conditional probability, now this is a very powerful Theorem where u know probability of certain events but u dont know the probability of some other events and using those certain events u can find other event probabilities


You all know about Titanic crash, we have a dataset of tis titanic crash where there are name of ppl along with certain features like Fare, ticket, Cabin, etc based on that we are trying to find out the Survival rate and here we can use <b>Bayes Theorem</b> where we are trying to find the probavility of Survival based on features such as if the person was Male, the class, age, cabin and fare and so on..

The reason its called <b>Naive Bayes</b> is bevause we make Naive assumption that the features such as Male, class, age cabin, fare etc are independent of each other. In reality some of these features might be dependent such as fare and cabin are kind of related but we assume for simplicity purpose that these arenot related hence it is called <b>Naive Bayes</b> and it is a simple assumption which reduces our calculation and makes the Algorithm simple yet effective


Naive Bayes is used in Email Spam detection, hand digit character recognization, whether predictions, face detections and news article categorization


# Coding Part

Download the titanic dataset from Kaggle

```python
import pandas as pd
```

```python
data = pd.read_csv('titanic2.csv')
data.head()
```

```python
data.shape
```

So we can see that some vairables doesnt haave any impact and it doesnt matter if we drop them, for example Name, it doesnt have any impact, so we will drop those variables which doesnt seem to be imprttant

```python
data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns'
          ,inplace=True)
data.head()
```

```python
X = data.drop(['Survived'], axis=1)
y = data['Survived']
```

Now we will turn the 'Sex' column into a numerical value using Dummy variable

```python
dummies = pd.get_dummies(data.Sex).astype(int)
dummies.head()
```

```python
X = pd.concat([X, dummies], axis=1)
X
```

```python
X.drop(['Sex'], axis=1, inplace=True)
X
```

Lets also find out if there is NaN value

```python
X.columns[X.isna().any()]
```

We can see that it tells us that the 'Age' column has some NaN value. lets chheck it

```python
X.Age.head(30)
```

We can see above that there is some NaN values.

Lets just fill the NaN value with the mean value of the Age column

```python
X.Age = X.Age.fillna(X.Age.mean())
X.head(30)
```

So we took care of the NaN value

### Train Test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

# Naive Bayes model

Now there are couple of Naive bayes classes but we are going to use 'GaussianNB'. u use this when ur data distribution is normal 

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
